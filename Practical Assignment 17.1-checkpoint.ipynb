```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier



file = r"C:\Users\rsreenivsan\Downloads\bank-full.csv"
df = pd.read_csv(file)
df.head()
```

```python
#drop day of contact as that may be irrelevant
df = df.drop(['day'], axis = 1)
```

```python
df.info()
```

```python
# One-hot encoding the categorical variables
bank_data_encoded = pd.get_dummies(df, columns=['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome'], drop_first=True)
```

```python
#Display first 10 rows
bank_data_encoded.head()
```

```python
#Standardize numerical features
from sklearn.preprocessing import StandardScaler

numerical_features = ['age', 'balance', 'duration', 'campaign', 'pdays', 'previous']
scaler = StandardScaler()
bank_data_encoded[numerical_features] = scaler.fit_transform(bank_data_encoded[numerical_features])
bank_data_encoded.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#visualizing box plots of all columns

plt.figure(figsize = (15,10))
for i, value in enumerate(numerical_features, 1):
    plt.subplot(4,2,i)
    sns.boxplot(y = bank_data_encoded[value])
    plt.title (value)
    

    
```

```python
#Data in campaign, balance, previous have some extreme outliers that may skew the model so capping some of them
#Using the definition of an outlier to determine the range to consider in 'balance' (1.5 IQR below and above Q1/Q3 quantile respectively)

previous_capped = bank_data_encoded['previous'].quantile(0.99)
campaign_capped = bank_data_encoded['campaign'].quantile(0.95)
Q1_quantile = bank_data_encoded['balance'].quantile(0.25)
Q3_quantile = bank_data_encoded['balance'].quantile(0.75)
IQR = Q3_quantile - Q1_quantile
lower_bal = Q1_quantile - 1.5*IQR 
upper_bal = Q3_quantile + 1.5*IQR

bank_data_encoded['campaign'] = bank_data_encoded['campaign'].clip(upper=campaign_capped)
bank_data_encoded['previous'] = bank_data_encoded['previous'].clip(upper=previous_capped)
bank_data_encoded['balance'] = bank_data_encoded['balance'].clip(upper=upper_bal, lower = lower_bal)

plt.figure(figsize = (15,10))
for i, value in enumerate(numerical_features, 1):
    plt.subplot(4,2,i)
    sns.boxplot(y = bank_data_encoded[value])
    plt.title (value)
    
#The box plots indicate a compressed range of values now without the outliers
```

```python
bank_data_encoded['y'] = bank_data_encoded['y'].map({'yes': 1, 'no': 0})
```

```python
#Splitting the data into test and train sets
X_train, X_test, y_train, y_test = train_test_split(bank_data_encoded.drop('y', axis = 1), bank_data_encoded.y, test_size=0.33, random_state=42)
```

```python
#Testing various models 

models = {"Logistic Regression": LogisticRegression(max_iter = 1000, random_state = 42),
         "Support Vector Machine": SVC(random_state = 42),
          "kNN":KNeighborsClassifier(n_neighbors = 5),
          "Decision Trees": DecisionTreeClassifier(random_state=0)
         }
results = {}
for name, model in models.items():
    #Training the model
    model.fit(X_train, y_train)
    
    #Predicting the test set
    y_pred = model.predict(X_test)
    
    #Evaluating the model
    accuracy = accuracy_score(y_test, y_pred)
    
    results[name] = (accuracy)
    

    
```

```python
print ("kNN accuracy:", results["kNN"], 
       "Logistic Regression accuracy:", results["Logistic Regression"], 
       "Decision trees accuracy:", results["Decision Trees"], 
       "SVM accuracy:", results["Support Vector Machine"])
```
